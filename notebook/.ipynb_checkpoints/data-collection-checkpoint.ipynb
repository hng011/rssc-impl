```python
import os
import glob
import shutil
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
raw_path_ds = "./datasets/NWPU-RESISC45/NWPU-RESISC45"
categories = os.listdir(raw_path_ds)
clean_path_ds = "./datasets/CLEAN_NWPU-RESISC45"
```

```python
len(categories)
```

# Data Understanding

## Create Clean Dataset Folder

```python
%%time

if not os.path.exists(clean_path_ds):
    os.mkdir(clean_path_ds)
    print("Successfully created a new dir for clean data")
else:
    print(f"{clean_path_ds} Already Exists")

for i, x in enumerate(categories):
    fn = os.listdir(os.path.join(raw_path_ds, x))
    for y in fn:
        if y.split(".")[-1] == "jpg":
            dest_path = os.path.join(clean_path_ds, x)
            if not os.path.exists(dest_path):
                os.makedirs(dest_path)
                
            shutil.copy2(os.path.join(raw_path_ds, x, y), os.path.join(dest_path, y))
```

## Create Sample Dataset Folder

```python
%%time

sample_amount = 500
dest_sample_path = f"./datasets/CLEAN_NWPU-RESISC45-SAMPLE{sample_amount}"

while 1:
    if not os.path.exists(dest_sample_path):
        os.mkdir(dest_sample_path)
        print("Successfully created a new dir for sample data")
    else:
        shutil.rmtree(dest_sample_path)
        break
    
for i, x in enumerate(categories):
    fl_raw = os.listdir(os.path.join(clean_path_ds, x))
    np.random.shuffle(fl_raw)
    fn = fl_raw[:sample_amount]
    
    n_path = os.path.join(dest_sample_path, x)
    for y in fn:
        if not os.path.exists(n_path):
            os.makedirs(n_path)
        shutil.copy2(os.path.join(clean_path_ds, x, y), os.path.join(n_path, y))
```

## Check

```python
import pandas as pd

total_img_each_cat = {
    "label": [],
    "total_data": []
}

for i, x in enumerate(os.listdir(dest_sample_path)):
    total_img_each_cat["label"].append(x)
    total_img_each_cat["total_data"].append(len(os.listdir(os.path.join(dest_sample_path, x))))
    
df = pd.DataFrame(total_img_each_cat)
df
```

```python
from PIL import Image
import matplotlib.pyplot as plt

clean_categories = os.listdir(dest_sample_path)
num_show = 9 # should be an odd num

plt.figure(figsize=(15, 10))
for i, label in enumerate(np.random.permutation(clean_categories)[:num_show]):
    img_path = os.path.join(dest_sample_path, label, np.random.choice(os.listdir(os.path.join(dest_sample_path, label))))
    img = Image.open(img_path)
    plt.subplot(int(np.sqrt(num_show)), int(np.sqrt(num_show)), i+1)
    plt.imshow(img)
    plt.title(label)
    # plt.axis("off")

plt.tight_layout()
plt.show()
```

# Data Preparation

## Data Splitting into Train and Test 

```python
import splitfolders

splitfolders.ratio("./datasets/CLEAN_NWPU-RESISC45-SAMPLE500",
                   output="./datasets/CLEAN_NWPU-RESISC45-SAMPLE500-SPLITTED8515",
                   seed=1, ratio=(.85, .0, .15),
                   group_prefix=None, move=False)
```

## Data Train Splitting into Train and Validation

### Prepare Images and Labels

```python
train_dir = "./datasets/CLEAN_NWPU-RESISC45-SAMPLE500-SPLITTED8515/train"

img_paths = glob.glob(os.path.join(train_dir, '*', "*.jpg"))
labels = [os.path.basename(os.path.dirname(p)) for p in img_paths]

print(f"""
len images: {len(img_paths)}
len labels: {len(labels)}
""")

print("======== VERIFYING ========")
for _ in range(10):
    rand_int = np.random.randint(0, len(img_paths))
    print(f"{img_paths[rand_int]} : {labels[rand_int]}")
```

### Splitting into Train and Validation

```python
data_train, data_val, label_train, label_val = train_test_split(
    img_paths,
    labels,
    test_size=0.15/0.85,
    stratify=labels,
    random_state=42
)

print(f"""
LEN DATA TRAIN: {len(data_train)} 
LEN LABEL TRAIN: {len(label_train)}

LEN DATA VAL: {len(data_val)}
LEN LABEL VAL: {len(label_val)}
""")
```

```python
len_ver = 10  

print("======== VERIFYING TRAIN DATA ========")
for _ in range(len_ver):
    rand_int = np.random.randint(0, len(data_train))
    print(f"{data_train[rand_int]} : {label_train[rand_int]}")
    
print("\n======== VERIFYING VAL DATA ========")
for _ in range(len_ver):
    rand_int = np.random.randint(0, len(data_val))
    print(f"{data_val[rand_int]} : {label_val[rand_int]}")    
```

### Saving the Stratified Train and Val Data

```python
def cp_file(img_data, img_labels, target_root):
    for p, l in zip(img_data, img_labels):
        target_dir = os.path.join(target_root, l)
        os.makedirs(target_dir, exist_ok=True)
        shutil.copy2(p, os.path.join(target_dir, os.path.basename(p)))
        
target_root_stratified = "./datasets/STRATIFIED_NWPU-RESISC45-500-TRAINVAL7015"

# Data Train        
cp_file(data_train, label_train, f"{target_root_stratified}/train")
 
# Data Val        
cp_file(data_val, label_val, f"{target_root_stratified}/val")
```
