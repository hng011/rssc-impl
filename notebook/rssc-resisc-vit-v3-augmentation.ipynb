# Data Preparation

```python
path_ds = "../input/stratified-nwpu-resisc45-500-trainval7015/STRATIFIED_NWPU-RESISC45-500-TRAINVAL7015"
```

```python
import datasets
 
def create_image_folder_dataset(root_path):
  """creates `Dataset` from image folder structure"""
 
  # get class names by folders names
  _CLASS_NAMES= os.listdir(root_path)
  # defines `datasets` features`
  features=datasets.Features({
                      "img": datasets.Image(),
                      "label": datasets.features.ClassLabel(names=_CLASS_NAMES),
                  })
  # temp list holding datapoints for creation
  img_data_files=[]
  label_data_files=[]
  # load images into list for creation
  for img_class in os.listdir(root_path):
    for img in os.listdir(os.path.join(root_path,img_class)):
      path_=os.path.join(root_path,img_class,img)
      img_data_files.append(path_)
      label_data_files.append(img_class)
  # create dataset
  ds = datasets.Dataset.from_dict({"img":img_data_files,"label":label_data_files},features=features)
  return ds
 
```

## Load Data

```python
train_ds = create_image_folder_dataset(f"{path_ds}/train")
train_ds
```

```python
val_ds = create_image_folder_dataset(f"{path_ds}/val")
val_ds
```

```python
classes = train_ds.features["label"].names
classes
```

## Preprocess

```python
from transformers import ViTFeatureExtractor
from tensorflow import keras
from tensorflow.keras import layers
 
model_id = "google/vit-base-patch16-224"

feature_extractor = ViTFeatureExtractor.from_pretrained(model_id)

augmentation_layers = tf.keras.Sequential([
    tf.keras.layers.RandomZoom(height_factor=0.1, width_factor=0.1),
    tf.keras.layers.RandomFlip("horizontal"),
])

def process(examples):
    aug_imgs = []

    for img in examples["img"]:
        img_tensor = tf.convert_to_tensor(np.array(img), dtype=tf.float32)

        # normalize to [0, 1] (tf expects this)
        img_tensor = img_tensor / 255.
        
        img_tensor = augmentation_layers(img_tensor)
        
        # Convert back to uint8 NumPy array (ViTFeatureExtractor expects this)
        img_aug = tf.clip_by_value(img_tensor * 255., 0, 255).numpy().astype(np.uint8)

        aug_imgs.append(img_aug)
    
    examples.update(feature_extractor(aug_imgs, ))
    return examples
```

```python
train_ds = train_ds.rename_column("label", "labels")
val_ds = val_ds.rename_column("label", "labels")
```

```python
train_ds_preprocessed = train_ds.map(process, batched=True)
train_ds_preprocessed
```

```python
val_ds_preprocessed = val_ds.map(process, batched=True) 
val_ds_preprocessed
```

## Convert DS to TF Datasets

```python
id2label = {str(i): label for i, label in enumerate(classes)}
label2id = {v: k for k, v in id2label.items()}
 
num_train_epochs = 15
batch_size = 10
learning_rate = 1e-5
weight_decay_rate = 1e-2
num_warmup_steps = 0
```

```python
id2label
```

```python
label2id
```

```python
from transformers import DefaultDataCollator
 
# Data collator that will dynamically pad the inputs received, as well as the labels.
data_collator = DefaultDataCollator(return_tensors="tf")
 
# converting our train dataset to tf.data.Dataset
train_data_augmentation = train_ds_preprocessed.to_tf_dataset(
   columns=["pixel_values"],
   label_cols=["labels"],
   shuffle=True,
   batch_size=batch_size,
   collate_fn=data_collator)
 
# converting our val dataset to tf.data.Dataset
val_data = val_ds_preprocessed.to_tf_dataset(
   columns=["pixel_values"],
   label_cols=["labels"],
   shuffle=True,
   batch_size=batch_size,
   collate_fn=data_collator)
```

## Apply Data Prefetch

```python
AUTOTUNE = tf.data.AUTOTUNE
train_data_augmentation = train_data_augmentation.prefetch(buffer_size=AUTOTUNE)
val_data = val_data.prefetch(buffer_size=AUTOTUNE)
```

# Modelling

```python
from transformers import TFViTForImageClassification, create_optimizer
import tensorflow as tf
 
# create optimizer wight weigh decay
num_train_steps = len(train_data_augmentation) * num_train_epochs
optimizer, lr_schedule = create_optimizer(
    init_lr=learning_rate,
    num_train_steps=num_train_steps,
    weight_decay_rate=weight_decay_rate,
    num_warmup_steps=num_warmup_steps,
)

def vit():
    # load pre-trained ViT model
    model = TFViTForImageClassification.from_pretrained(
        model_id,
        num_labels=len(classes),
        id2label=id2label,
        label2id=label2id,
        ignore_mismatched_sizes=True
    ) 
     
    # compile model
    model.compile(
        optimizer=optimizer,
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        metrics=["accuracy"]
    )
    
    model.summary()
    print(f"\n##### Optimizer Func Information\n{model.optimizer.get_config()}")
    print(f"\n##### Loss Func Information\n{model.loss.get_config()}")
    
    return model
```

## 15 Epochs with Augmentation

```python
aug_vit_p16 = vit()
```

```python
%%time

history_aug_vit_p16 = aug_vit_p16.fit(
    train_data_augmentation,
    validation_data=val_data,
    epochs=num_train_epochs,
)
```

```python
# base_vit_p16.save("base_vit_p16_epoch0010.keras")
aug_vit_p16.save_pretrained("./aug_vit_p16_in1k")
joblib.dump(history_aug_vit_p16.history, "history_aug_vit_p16_in1k_epoch0015.joblib")
```
